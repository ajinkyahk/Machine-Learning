## Difference Between LOC and ILOC

**Let's Create a Sample Dataframe**

```python
sample_df = pd.DataFrame({
    'gender' : ['M', 'F', 'M', 'M', 'F'],
    'grade' : ['A', 'A', 'B', 'B', 'A'],
    'marks' : [22, 21, 12, 14, 20],
    'id' : ['A101', 'A102', 'A103', 'A104', 'A105']
})

sample_df
```

## Now, when we define a dataframe by default the index is a range of numbers. Let's see what happens if try to slice the dataframe using both <font color='orange'>loc</font> and <font color='green'>iloc</font>

```python
sample_df.loc[2:4]
```

```python
sample_df.iloc[2:4]
```

- **When we try to slice the dataframe using the<span style='background:yellow'> loc</span> function on <span style='background:yellow'>range(2 to 4)</span> it first finds out the index with a <span style='background:yellow'>label 2</span> and goes till it finds the index with a <span style='background:yellow'>label 4</span>.**

- **In the other hand, when we try to slice the dataframe using <span style='background:yellow'>iloc</span> function on <span style='background:yellow'>range(2 to 4)</span> it statrts with the <span style='background:yellow'>starting point index 2</span> and goes till <span style='background:yellow'>end point -1 which is 3</span>.**

**Let's verify this by sorting the dataframe by marks column.**

```python
sample_df = sample_df.sort_values(by='marks')
sample_df
```

## Now that we have sorted the dataframe, the indices are shuffled accordingly. Let's try to slice the dataframe again.

```python
sample_df.loc[2:4]
```

```python
sample_df.iloc[2:4]
```

```python
sample_df.loc[3:1]
```

- **You can see that when we tried to slice the dataframe using the loc function it finds out the name of start and end point and slie the data whereas iloc gives the 2nd and 3rd value of the dataframe.**

**Let's change the index of the dataframe to a categorical variable.**

```python
sample_df.set_index('id', inplace=True)
sample_df
```
