# Reading Big CSV Files in Python

```python
import pandas as pd
```

### READ THE FIRST 100 ROWS

```python
data_100 = pd.read_csv('dataset/big_mart_sales.csv', nrows=101)
```

```python
data_100.head()
```

```python
data_100.tail()
```

```python
data_100.shape
```

### READ THE NEXT 100 ROWS

```python
data_100_200 = pd.read_csv('dataset/big_mart_sales.csv', skiprows=101, nrows=100)
```

```python
data_100_200.head()
```

```python
data_100_200.tail()
```

```python
data_100_200.shape
```
