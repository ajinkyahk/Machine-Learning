# Sorting the Data

**We will see how to sort the data based on a single column and on multiple columns. Also we will see how to reset index after sorting the data.**

```python
# import pandas library
import pandas as pd
```

## Create a Simple Dataframe
- We will create a simple data of students that will contain roll_no, name, grade, marks and city.

```python
# sample dataframe

data_frame = pd.DataFrame({
    'roll_no' : [102, 101, 104, 103, 105],
    'name' : ['Arvind', 'Rahul', 'Prateek', 'Piyuesh', 'Kartik'],
    'grade' : ['B', 'B', 'A', 'C', 'A'],
    'marks' : [15, 15, 20, 4, 22],
    'city': ['Gurugram', 'Delhi', 'Delhi', 'Gurugram', 'Hyderabad']
})
data_frame
```

## SORT THE DATAFRAME BY THE GRADES OF THE STUDENTS

- Now, we need to sort the students based on their grades. We will use the sort_values function and the parameter by we will pass the column grade.

```python
# sort the data

data_frame.sort_values(by=['grade'])
```

```python
# sort the data

data_frame.sort_values(by=['roll_no'])
```

- Now, that we have sorted data based on the grades. You can see that some of the students with higher marks are ranked lower even having the same marks. So, we want to sort hte data based on both grades and marks.
- Also, the grades should be in ascending order and marks should be in descending order.

---

## SORT THE DATAFRAE BY GRADE AND MARKS OF THE STUDENTS

- To sort the values in the descending order, we need to set parameter ascending = False.
- In the by parameter pass the list of columns on which we want to sort and for the ascending parameter pass the boolean list True for ascending and False for descending.

```python
# sort the dataframe
data_frame.sort_values(by=['grade', 'marks'], ascending=[True, False])
```

# Now let's see the dataframe

```python
data_frame
```

## Now, you can observe that the sorted state of the dataframe is not saved. Use the parameter inplace=True to save the sorted state.

```python
# save the sorted state by using inplace=True

data_frame.sort_values(by=['grade', 'marks'], ascending=[True, False], inplace=True)
```

```python
data_frame
```

## Now, we can see that the index is also shuffled according to the sorting. If we want to reset the index we use reset_index function.

```python
data_frame.reset_index()
```

## Now it has created another column index which is the previous inde. If you want to remove this just past the parameter drop=True and also inplace= True to save the state.

```python
data_frame.reset_index(drop=True, inplace=True)
data_frame
```
