#### How to view the top and bottom rows of the data?

- Use head function to view top n rows.
- Use tail function to view bottom n rows.

```
# view top 2 rows 
data.head(2)
```

```
# view bottom rows

data.tail(2)
```

#### How to select rows in a particular range?

```
data.head()
```

```
#select the data from range 10-15
data[10:15]
```

```
# range with step 2
data[20:30:2]
```

```
#range with negative index and step 2
data[-10::2]
```

#### How to select rows by position?
- When we are using iloc, we need to specify the rows and columns by their posittion.

```
#select specific rows by index number
data.iloc[[1,5,2,4,6,14]]
```

#### How to select the specific rows and columns from the data using their positon?
- In the iloc function pass the first list as the order of rows by their index and pass the second list as the order of columns.

```
data.iloc[[1,4,5,2], [1,3,5]]
```

```
data.iloc[10:15]
```
